### Importación de datos



```python
import pandas as pd

pd.options.display.float_format = '{:,.0f}'.format

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
#Calculamos el total de facturacion
facturacion_total_tienda1 = tienda1["Precio"].sum()
facturacion_total_tienda2 = tienda2["Precio"].sum()
facturacion_total_tienda3 = tienda3["Precio"].sum()
facturacion_total_tienda4 = tienda4["Precio"].sum()

print("La facturación total de la tienda 1 es:", facturacion_total_tienda1)
print("La facturación total de la tienda 2 es:", facturacion_total_tienda2)
print("La facturación total de la tienda 3 es:", facturacion_total_tienda3)
print("La facturación total de la tienda 4 es:", facturacion_total_tienda4)

```

```python
#Grafica de ingresos totales por cada tienda.

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ingresos_totales = [facturacion_total_tienda1, facturacion_total_tienda2, facturacion_total_tienda3, facturacion_total_tienda4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

colores =  ["green", "blue", "yellow", "red"]
plt.figure(figsize=(5, 4))
plt.bar(tiendas, ingresos_totales, color=colores)

plt.xlabel("Tiendas")
plt.ylabel("Ingresos totales ($)")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)



plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("${x:,.0f}"))
plt.title("Ingresos totales por tienda 01/2022 - 03/2023")
plt.show()
```

Se evidencia que la tienda numero 1 reporta mayor cantidad de ingresos sumando todas las categorias de productos, y con menor reporte de venta es la tienda numero 4.

# 2. Ventas por categoría

```python
import pandas as pd

data = {
    'tienda': [171, 284, 312, 448, 182, 324, 173, 465],
    'tienda2': [181, 275, 305, 422, 224, 313, 197, 442],
    'tienda3': [177, 277, 278, 451, 177, 315, 185, 499],
    'tienda4': [201, 277, 254, 451, 170, 338, 187, 480]
}

index = ['Artículos para el hogar', 'Deportes y diversión', 'Electrodomésticos', 'Electrónicos', 'Instrumentos musicales', 'Juguetes', 'Libros', 'Muebles']

df = pd.DataFrame(data, index=index)
df.index.name = 'Categoría del Producto'

display(df)
```

Analizando las grafica de lineas nos indica que las categorias mas populares de las tiendas son: Muebles y Electronicos, las categorias de Juguetes y electrodomesticos estas en tercera y cuarta posicion.  


# 3. Calificación promedio de la tienda


```python
calif_prom_t1 = tienda1["Calificación"].mean()
calif_prom_t2 = tienda2["Calificación"].mean()
calif_prom_t3 = tienda3["Calificación"].mean()
calif_prom_t4 = tienda4["Calificación"].mean()

print(f"Calificación promedio Tienda 1: {calif_prom_t1:.2f}")
print(f"Calificación promedio Tienda 2: {calif_prom_t2:.2f}")
print(f"Calificación promedio Tienda 3: {calif_prom_t3:.2f}")
print(f"Calificación promedio Tienda 4: {calif_prom_t4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Calcular las calificaciones promedio (asumiendo que ya tienes tienda1, tienda2, tienda3, tienda4)
calif_prom_t1 = tienda1["Calificación"].mean()
calif_prom_t2 = tienda2["Calificación"].mean()
calif_prom_t3 = tienda3["Calificación"].mean()
calif_prom_t4 = tienda4["Calificación"].mean()

# Datos para el gráfico
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [calif_prom_t1, calif_prom_t2, calif_prom_t3, calif_prom_t4]

# Crear el gráfico de marcadores
plt.plot(tiendas, calificaciones, marker='o', linestyle='-', color='b')

# Añadir etiquetas y título
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")
plt.title("Calificación Promedio por Tienda")

# Mostrar valores en los marcadores
for i, txt in enumerate(calificaciones):
    plt.annotate(f'{txt:.2f}', (tiendas[i], calificaciones[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

Al realizar la calificacion promedio de satisfaccion del cliente, observamos que la tienda numero 1, necesita una mayor atencion en servicio al cliente, siendo esta la de mayor ventas, por esta razon se recomendable capacitacion al personal sobre, servicio al cliente. ya que si se mejora en este aspecto se puede traducir en mayores ventas y adquisicion de nuevos clientes.

# 4. Productos más y menos vendidos

```python
import pandas as pd

# URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Carga de datos
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Lista para resultados
destacados = []

# Procesamiento por tienda
for tienda_df, nombre in zip(tiendas, nombres):
    ventas = tienda_df['Producto'].value_counts()

    top_5 = ventas.head(5)
    bottom_5 = ventas.tail(5)

    for producto, cantidad in top_5.items():
        destacados.append({
            'Tienda': nombre,
            'Tipo': 'Más Vendido',
            'Producto': producto,
            'Cantidad': cantidad
        })

    for producto, cantidad in bottom_5.items():
        destacados.append({
            'Tienda': nombre,
            'Tipo': 'Menos Vendido',
            'Producto': producto,
            'Cantidad': cantidad
        })

# Crear DataFrame final
df_destacados = pd.DataFrame(destacados)

# Mostrar resultados
print("Top 5 productos más y menos vendidos por tienda:\n")
print(df_destacados)
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio_tienda1 = tienda1['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio tienda 1: {costo_envio_promedio_tienda1}")
print(f"Costo de envío promedio tienda 2: {costo_envio_promedio_tienda2}")
print(f"Costo de envío promedio tienda 3: {costo_envio_promedio_tienda3}")
print(f"Costo de envío promedio tienda 4: {costo_envio_promedio_tienda4}")

# Datos de costo de envío promedio de cada tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos_envio = [costo_envio_promedio_tienda1, costo_envio_promedio_tienda2, costo_envio_promedio_tienda3, costo_envio_promedio_tienda4]

# Crear la gráfica de barras para el costo de envío promedio
plt.figure(figsize=(8, 6))
plt.bar(tiendas, costos_envio, color='lightcoral')
plt.xlabel('Tiendas')
plt.ylabel('Costo de Envío Promedio')
plt.title('Costo de Envío Promedio por Tienda')

# Agregar los valores de costo de envío encima de cada barra
for i, v in enumerate(costos_envio):
    plt.text(i, v, f'{v:.2f}', ha='center', va='bottom')

plt.show()
```
